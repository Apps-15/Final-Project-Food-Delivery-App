```python
class FoodItem:
    def __init__(self, name, quantity, price, discount, stock):
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock


class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

    def place_order(self, selected_items):
        total_price = 0
        ordered_items = []
        for index in selected_items:
            if index >= 0 and index < len(menu):
                item = menu[index]
                if item.stock > 0:
                    total_price += item.price
                    ordered_items.append(item)
                    item.stock -= 1
        if ordered_items:
            self.orders.append(ordered_items)
            return total_price
        return 0


class Admin:
    def add_food_item(self, name, quantity, price, discount, stock):
        food_item = FoodItem(name, quantity, price, discount, stock)
        menu.append(food_item)

    def edit_food_item(self, index, name, quantity, price, discount, stock):
        if index >= 0 and index < len(menu):
            food_item = menu[index]
            food_item.name = name
            food_item.quantity = quantity
            food_item.price = price
            food_item.discount = discount
            food_item.stock = stock

    def view_food_items(self):
        for index, item in enumerate(menu):
            print(f"{index}. {item.name} - {item.quantity} - INR {item.price}")

    def remove_food_item(self, index):
        if index >= 0 and index < len(menu):
            del menu[index]


menu = []
admin = Admin()
user = User("John Doe", "1234567890", "john@example.com", "123 Main St", "password")

# Admin adds food items
admin.add_food_item("Tandoori Chicken", "4 pieces", 240, 0, 10)
admin.add_food_item("Vegan Burger", "1 Piece", 320, 10, 5)
admin.add_food_item("Truffle Cake", "500gm", 900, 5, 8)

# User places an order
selected_items = [1, 2]
total_price = user.place_order(selected_items)
if total_price > 0:
    print("Order placed successfully!")
    print(f"Total price: INR {total_price}")
else:
    print("No items available for ordering.")

# Admin views food items
admin.view_food_items()
```
